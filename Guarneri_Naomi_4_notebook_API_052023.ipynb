```python
import json
```

```python
dic=json.loads(data)
```

```python
foods = dic['hints']
foods
```

```python
foods[0]['food'].keys()
```

```python
import pandas as pd
df = pd.DataFrame(columns=['foodId', 'label','category','foodContentsLabel','image'])
```

```python
for food in foods :
    foodId=food['food']['foodId']
    label=food['food']['label']
    category=food['food']['category']
    foodContentsLabel = food['food']['foodContentsLabel'] if 'foodContentsLabel' in food['food'].keys() else 'None'
    image = food['food']['image'] if 'image' in food['food'].keys() else 'None'
    df.loc[len(df)] = [foodId,label,category,foodContentsLabel,image]
```
